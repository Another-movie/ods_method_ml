## Seaborn

Seaborn — высокоуровневое API на базе matlib

Распределяем оценки критиков с помощью __`distplot`__. По дефолту на графике отображается гистограмма и kernel density estimation (kde)

```python
sns.distplot(data.Critic_Score)
```

Для того, чтобы посмореть на взаимосвязь двух численных признаков, используем – __`jointplot`__, это гибрид `scatter plot` и `histogram`

```python
%config InlineBackend.figure_format = 'png' # lagi pri svg
sns.jointplot(x='Critic_Score', y='User_Score', data=data, kind='scatter')
```

```python
top_platforms = data.Platform.value_counts().sort_values(ascending=False).head(5).index.values
sns.boxplot(y="Platform", x="Critic_Score",
           data=data[data.Platform.isin(top_platforms)], orient="h")
```

Box plot состоит из коробок, усиков и точек. Коробка показывает интерквантильный размах распределения, то есть 25%(Q1) и 75%(Q3) процентили. Черта внутри коробки - медиана. Усы отображают весь разброс точек кроме выбросов, то есть минимальные и максимальные значения, которые попадают в промежуток `(Q1 - 1.5*IQR, Q3 + 1.5*IQR)`, где `IQR = Q3 - Q1`.
Точками на графике обозначаются выбросы (`outliers`) - те значения, которые не входят в промежуток значений, заданный усами графика.

__`heat map`__ – позволяет посмотреть на распределние какого-то численного признака по двум категориальным

```python
platform_genre_sales = data.pivot_table(
                        index='Platform',
                        columns='Genre',
                        values='Global_Sales',
                        aggfunc=sum).fillna(0).applymap(float)
```

```python
sns.heatmap(platform_genre_sales, annot=True, fmt='.1f', linewidths=.5)
```
